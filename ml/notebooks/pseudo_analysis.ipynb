```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme(style="darkgrid")
plt.rcParams["figure.figsize"] = (20,5)
```

```python
df = pd.read_csv('../data/pseudo_parallel/temp/formality/formality_train_cross_predict_paraphrases.csv')
```

```python
df
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
sns.histplot(df, x="pred_formality_orig", ax = ax1)
sns.histplot(df, x="pred_formality_paraphrased", ax = ax2)
```

```python
df['formality_original_bucket'] = 'low'
df.loc[df['pred_formality_orig']>0.1,'formality_original_bucket'] = 'skip'
df.loc[df['pred_formality_orig']>0.2,'formality_original_bucket'] = 'mid'
df.loc[df['pred_formality_orig']>0.8,'formality_original_bucket'] = 'skip'
df.loc[df['pred_formality_orig']>0.9,'formality_original_bucket'] = 'high'


df['formality_paraphrase_bucket'] = 'low'
df.loc[df['pred_formality_paraphrased']>0.1,'formality_paraphrase_bucket'] = 'skip'
df.loc[df['pred_formality_paraphrased']>0.2,'formality_paraphrase_bucket'] = 'mid'
df.loc[df['pred_formality_paraphrased']>0.8,'formality_paraphrase_bucket'] = 'skip'
df.loc[df['pred_formality_paraphrased']>0.9,'formality_paraphrase_bucket'] = 'high'
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
sns.histplot(df, x="formality_original_bucket", ax=ax1)
sns.histplot(df, x="formality_paraphrase_bucket", ax=ax2)
```

```python
sns.heatmap(df.groupby(['formality_original_bucket','formality_paraphrase_bucket']).agg({'text': 'count'}).reset_index().pivot('formality_original_bucket', 'formality_paraphrase_bucket')/1000, annot=True)
```

```python
(df['pred_formality_orig']-df['pred_formality_paraphrased']).hist()
```

```python
df['diff_bucket'] = 'high'
df.loc[(df['pred_formality_orig']-df['pred_formality_paraphrased'])>-0.25,'diff_bucket'] = 'same'
df.loc[(df['pred_formality_orig']-df['pred_formality_paraphrased'])>0.25,'diff_bucket'] = 'low'
```

```python
df['text_tokens'] = df['text'].str.split(' ').apply(len)
df['paraphrased_tokens'] = df['paraphrased'].str.split(' ').apply(len)
df['text_dots']= df['text'].apply(lambda x: bool(sum([x.find(i)>=0 for i in ['..','!!','#',':','_','(',')','--']])) )
df['paraphrased_dots']= df['paraphrased'].apply(lambda x: bool(sum([x.find(i)>=0 for i in ['..','!!','#',':','_','(',')','--']])) )
```

```python
# df.groupby('text_tokens').agg({'text':'count'})
# df.groupby('paraphrased_tokens').agg({'text':'count'})
```

```python
len(df[(~df['text_dots'])&(~df['paraphrased_dots'])&(df['text_tokens']>=5)&(df['paraphrased_tokens']>=5)&(df['formality_paraphrase_bucket']!='skip')&(df['formality_original_bucket']!='skip')&(df['formality_paraphrase_bucket']!=df['formality_original_bucket'])]), len(df)
```

```python
df = df[(~df['text_dots'])&(~df['paraphrased_dots'])&(df['text_tokens']>=5)&(df['paraphrased_tokens']>=5)&(df['formality_paraphrase_bucket']!='skip')&(df['formality_original_bucket']!='skip')&(df['formality_paraphrase_bucket']!=df['formality_original_bucket'])]
df = df.sample(frac=1).reset_index(drop=True)
train_samples = int(len(df)*0.8)
train_df = df[:train_samples]
dev_df = df[train_samples:]
```

```python
sns.heatmap(train_df.groupby(['formality_original_bucket','formality_paraphrase_bucket']).agg({'text': 'count'}).reset_index().pivot('formality_original_bucket', 'formality_paraphrase_bucket')/1000, annot=True)
plt.show()
sns.heatmap(dev_df.groupby(['formality_original_bucket','formality_paraphrase_bucket']).agg({'text': 'count'}).reset_index().pivot('formality_original_bucket', 'formality_paraphrase_bucket')/1000, annot=True)
```

```python
train_df1 = train_df.filter(['text','formality_original_bucket','paraphrased','formality_paraphrase_bucket'])
dev_df1 = dev_df.filter(['text','formality_original_bucket','paraphrased','formality_paraphrase_bucket'])

train_df1.columns = ['input','input_formality','output','output_formality']
dev_df1.columns = ['input','input_formality','output','output_formality']

train_df2 = train_df1.copy()
dev_df2 = dev_df1.copy()
train_df2.columns = ['output','output_formality','input','input_formality']
dev_df2.columns = ['output','output_formality','input','input_formality']

train_df2 = train_df2.filter(['input','input_formality','output','output_formality'])
dev_df2 = dev_df2.filter(['input','input_formality','output','output_formality'])

train_df3 = pd.concat([train_df1, train_df2])
dev_df3 = pd.concat([dev_df1, dev_df2])

```

```python
sns.heatmap(train_df3.groupby(['input_formality','output_formality']).agg({'input': 'count'}).reset_index().pivot('input_formality', 'output_formality')/1000, annot=True)
plt.show()
sns.heatmap(dev_df3.groupby(['input_formality','output_formality']).agg({'input': 'count'}).reset_index().pivot('input_formality', 'output_formality')/1000, annot=True)
```
